```python
data1=data
```

```python
column_list=data.columns[3:]
```

```python
column_list
```

```python
# if df_1["복리후생비(연결)(천원)"]
from tqdm import tqdm
import numpy as np
for i in tqdm(np.arange(0,len(column_list),2)):
    for j in range(data1.shape[0]):
        if np.isnan(data1[column_list[i]][j]) :
            data1[column_list[i]][j]=data1[column_list[i+1]][j]
```
