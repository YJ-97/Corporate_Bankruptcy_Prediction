```python
import pandas as pd
```

```python
사업보고서 = pd.read_csv('./datasets/연간보고서09_18.csv', encoding='cp949')
사업보고서.dropna(inplace=True)
사업보고서.isna().sum()
사업보고서
```

```python
# 회계년도리스트 = 사업보고서.회계년도.str.split('-')
# 사업보고서.회계년도 = '20' + 회계년도리스트.str.get(1)
# 사업보고서
```

```python
회계년도리스트 = 사업보고서.회계년도.str.split('/')
회계년도리스트
사업보고서.회계년도 = 회계년도리스트.str.get(0)
사업보고서
```

```python
사업보고서['운전자본'] = 사업보고서['유동자산']-사업보고서['유동부채']
사업보고서['자기자본'] = 사업보고서['자산']-사업보고서['부채']

사업보고서['운전자산총자본비율'] = (사업보고서['운전자본'] / 사업보고서['자산']) * 100
사업보고서['이익잉여금총자산비율'] = (사업보고서['이익잉여금'] / 사업보고서['자산']) * 100
사업보고서['총자산이익율'] = (사업보고서['영업이익'] / 사업보고서['자산']) * 100
사업보고서['시장가부채비율'] = (사업보고서['자산'] / 사업보고서['부채']) * 100
사업보고서['매출액회전율'] = (사업보고서['매출액'] / 사업보고서['자산']) * 100
```

```python
사업보고서_기본재무비율 = 사업보고서[['회사명', '회계년도', '운전자산총자본비율', '이익잉여금총자산비율', '총자산이익율', '시장가부채비율', '매출액회전율']]
사업보고서_기본재무비율
```

```python
사업보고서_기본재무비율.to_csv('./datasets/사업보고서_기본재무비율09_18.csv', index=None, encoding='CP949')
```
