```python
import pandas as pd
import numpy as np
```

```python
all = pd.read_csv('./preprocessed_data/all_final.csv', encoding='cp949')
all_ = all[all['부실기업여부']==1]
all_ = all_[['회사명', '회계년도', '부실기업여부']]
all_
```

```python
시총 = pd.read_csv('시총.csv', encoding='CP949')
시총 = 시총[['회사명', '회계년도', '시총', '상장협 산업분류(중분류)']]
회계년도리스트 = 시총['회계년도'].str.split('-')
시총['회계년도'] = ('20' + 회계년도리스트.str.get(1))
시총.회사명 = 시총.회사명.str.replace('\(주\)', '')
시총.회계년도 = 시총.회계년도.astype(int)
시총
```
